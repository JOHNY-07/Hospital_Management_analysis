```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv("patients.csv")
df1
```

```python
df2=pd.read_csv("doctors.csv")
df2
```

```python
df3=pd.read_csv("appointments.csv")
df3
```

```python
df_merge_1=pd.merge(df1,df3,on="patient_id")
df_merge_1
```

```python
df=pd.merge(df_merge_1,df2,on="doctor_id")
df
```
